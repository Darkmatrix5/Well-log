```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('log_data/log.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
import seaborn as sns
sns.heatmap(df.isnull())
```

```python
fig,ax=plt.subplots(figsize=(4,8))

ax.plot(df['DCAL'],df['Depth'],label='Calliper')

ax.set_ylim(5000,0)
ax.legend(loc='upper right')
ax.grid()
```

```python
fig,ax=plt.subplots(figsize=(4,8))

ax.plot(df['SP'],df['Depth'],label='SP')
ax.plot(df['GR'],df['Depth'],label='GR')

ax.axvline(x=-100,c='r')
ax.axhspan(3800,4800,alpha=0.2,color='grey',label='Shale Bed')

# calculating volume of shale using gamma ray loh
gr_max = max(df['GR'])
gr_min = min(df['GR'])

df['Vshale'] = (df['GR']-gr_min)*100/(gr_max-gr_min)
ax.plot(df['Vshale'],df['Depth'],label="Vshale")

ax.set_ylim(5000,0)
ax.legend(loc='upper right')
ax.grid()
```

```python
fig,ax=plt.subplots(figsize=(4,10))

ax.plot(df['DPOR'],df['Depth'],label='Density_por')
ax.plot(df['CNLS'],df['Depth'],label='Neutrin_por')

ax.axhspan(1700,2500,alpha=0.2,color='yellow',label='expected Gas zone')
ax.axhspan(2500,3580,alpha=0.2,color='brown',label='expected Oil zone')

ax.set_ylim(5000,0)
ax.legend(loc='upper right')
ax.grid()
```

```python
fig,ax=plt.subplots(figsize=(4,10))

ax.plot(df['RILD'],df['Depth'],label='RILD')
ax.plot(df['RLL3'],df['Depth'],label='RLL3')

ax.axhspan(1700,2500,alpha=0.2,color='yellow',label='expected Gas zone')
ax.axhspan(2500,3580,alpha=0.2,color='brown',label='expected Oil zone')

ax.set_ylim(5000,0)
ax.legend(loc='upper right')
ax.grid()
```

```python
x = df['RILD']
x=np.array(x)
x[x>300]=300
# idx = np.argwhere(x > 1000)
y = df['RLL3']
y=np.array(y)
y[y>300]=300

# x = x.bfill()
```

```python
fig,ax=plt.subplots(figsize=(4,10))

ax.plot(x,df['Depth'],label='RILD')
ax.plot(y,df['Depth'],label='RLL3')

ax.axhspan(1700,2400,alpha=0.2,color='yellow',label='expected Gas zone')
ax.axhspan(4100,4800,alpha=0.2,color='brown',label='expected Oil zone')

ax.set_ylim(5000,0)
```

```python
fig,ax = plt.subplots(1,4,figsize=(12,8))

ax[0].plot(df['DCAL'],df['Depth'],label='Calliper')
ax[0].set_ylim(0,5000)
ax[0].legend(loc='upper right')
ax[0].grid()

ax[1].plot(df['SP'],df['Depth'],label='SP')
ax[1].plot(df['GR'],df['Depth'],label='GR')
ax[1].axvline(x=-100,c='r')
ax[1].axhspan(3800,4800,alpha=0.2,color='grey',label='Shale Bed')
ax[1].set_ylim(5000,0)
ax[1].legend(loc='upper right',prop = { "size": 6 })
ax[1].grid()

ax[2].plot(df['DPOR'],df['Depth'],label='Density_por')
ax[2].plot(df['CNLS'],df['Depth'],label='Neutrin_por')
ax[2].axhspan(1700,2500,alpha=0.2,color='#11FFF8',label='expected Gas zone')
ax[2].axhspan(2500,3580,alpha=0.2,color='#952803',label='expected Oil zone')
ax[2].set_ylim(5000,0)
ax[2].legend(loc='upper right',prop = { "size": 6 })
ax[2].grid()

ax[3].plot(x,df['Depth'],label='RILD')
ax[3].plot(y,df['Depth'],label='RLL3')
ax[3].axhspan(1700,2500,alpha=0.2,color='#11FFF8',label='High deep resistivity')
ax[3].axhspan(3800,4800,alpha=0.2,color='grey',label='high shallow resistivity')
ax[3].set_ylim(5000,0)
ax[3].legend(loc='upper right',prop = { "size": 6 })
ax[3].grid()
```
